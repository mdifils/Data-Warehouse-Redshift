# Data quality Check

## Check if artist_id is unique

```python
%%sql
SELECT count(artist_id), count(distinct artist_id)
FROM 
(
    SELECT DISTINCT
        artist_id,
        artist_name AS name,
        artist_location AS location,
        artist_latitude AS lattitude,
        artist_longitude AS longitude
    FROM staging_songs
    WHERE artist_id IS NOT NULL
);
```

# Check duplicated artist_id

```python
%%sql
SELECT artist_id, COUNT(*)
FROM
(
    SELECT DISTINCT
        artist_id,
        artist_name AS name,
        artist_location AS location,
        artist_latitude AS lattitude,
        artist_longitude AS longitude
    FROM staging_songs
    WHERE artist_id IS NOT NULL
)
GROUP BY artist_id
HAVING COUNT(*) > 1
ORDER BY artist_id
LIMIT 20;
```

## Check why artist_id is not unique
See README file for explanation

```python
%%sql
SELECT a.*
FROM 
(
    SELECT DISTINCT
        artist_id,
        artist_name AS name,
        artist_location AS location,
        artist_latitude AS lattitude,
        artist_longitude AS longitude
    FROM staging_songs
    WHERE artist_id IS NOT NULL
) a
JOIN 
(
SELECT artist_id, COUNT(*)
FROM
(
    SELECT DISTINCT
        artist_id,
        artist_name AS name,
        artist_location AS location,
        artist_latitude AS lattitude,
        artist_longitude AS longitude
    FROM staging_songs
    WHERE artist_id IS NOT NULL
)
GROUP BY artist_id
HAVING count(*) > 1 
) b
ON a.artist_id = b.artist_id
ORDER BY artist_id
LIMIT 20;
```
