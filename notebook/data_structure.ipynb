```python
import boto3
from getpass import getpass
```

# Getting access and secret keys

```python
print("Enter your ACCESS KEY: ")
KEY = getpass()
print("Enter your SECRET KEY: ")
SECRET = getpass()
```

# Creating a connection to AWS S3 and accessing "udacity-dend" bucket

```python
s3 = boto3.resource('s3',
                    region_name="us-west-2",
                    aws_access_key_id=KEY,
                    aws_secret_access_key=SECRET
                   )
sampleDbBucket =  s3.Bucket("udacity-dend")
```

# Accessing one "song_data" file

```python
obj_list = list(sampleDbBucket.objects.filter(Prefix='song_data/A/B/C/'))
obj = obj_list[0]
obj
```

```python
import pandas as pd
df_song_data = pd.read_json('s3://udacity-dend/song_data/A/B/C/TRABCAS128F14A25E2.json',  lines=True)
df_song_data
```
